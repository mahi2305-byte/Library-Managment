```python
student_count = 0
class Accounts:
    student_count = 0
    
    def __init__ (self , name , balance , password):
        self.name = name
        self.balance = balance
        self._password = password 
        
    def deposit(self , amount):
        if amount > 0 :
            self.balance += amount 
            print(f"Deposited: {amount} , New balance: {self.balance}")
            
        else :
            print("invalid value")
        
    def withdraw(self, amount):
        if amount > self.balance:
            print("insufficient balance")
            
        elif amount > 0 and amount <= self.balance:
            self.balance -= amount 
            print (f"Withdrawed : {amount} , New balance : {self.balance}")
            
        else :
            print("insufficient amount")
            
    def get_balance(self):
        print ("total balance", self.balance)
        
    def check_password(self , password):
        if password == self._password:
            print("ACESS GRANTED")
            return True
        else:
            
            print ("ACESS DENIED WRONG PASSWORD")
            return False
            
    @classmethod #used as decorator 
    def get_count(cls):
        return cls.student_count
    
    
    
```

```python

user1 = Accounts("mahi" , 100 , "1234")
user2 = Accounts("garg" , 200 , "2305")
    
```

```python
users = [user1 , user2 ]
```

```python
def login():
    username = input("enter username")
    password = input("enter password")
    
    for user in users:
        if username == user.name :
            if user.check_password(password):
                return user 
            
def get_menu(user):
                
    print (f"Welcome {user.name}!")
    print("1.Get Balance")
    print("2.Withdraw")
    print("3. Deposit")
    print("4. Exit")
                    
                               
while True:
    user = login()
    # print(user)
    if user:
        while True:
            get_menu(user)
                
            
            choice = int(input("enter choice"))
            if choice == 1:
                user.get_balance()
                    
            elif choice == 2:
                amount = float(input("enter amount to withdraw"))
                user.withdraw(amount)
                    
            elif choice == 3:
                
                amount = float(input("enter amount to deposite"))
                user.deposit(amount)
                    
            else:
                break 
                
                

                    
            
```
