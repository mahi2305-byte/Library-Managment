```python
from typing import List 

def user_input():
        student = input("Enter student name")
        marks = []
        for i in range (1,6):
                while True:
                        
                        mark = float(input(f" Enter marks for subject {i}:"))
                        if 0 <= mark <= 100:
                                marks.append(mark)
                                break 
                        else:
                                print("Invalid marks , enter marks between 0 - 100")
                        
                        
            
        return student,marks
        
        
def calculate_avg( marks : list[float]):
         avg = sum(marks)/len(marks)
         
         return avg 
       

        
        
        
def assign_grade(avg):
        if avg >= 90:
                return "A+"
        elif avg >= 80:
                return "A"
        elif avg >= 70:
                return "B"
        elif avg >= 60:
                return "C" 
        elif avg >= 50:
                return "D"
        else:
                return "F"
        
        

def pass_or_fail(avg):
        if avg>= 50:
                return "Pass"
        else:
                return "Fail"
        
def display_result(name , average , Grade , status):
        print("---Result---")
        print(f"Student name: {name}")
        print(f"Average :{average}")
        print(f"Grade : {Grade}")
        print(f"Status : {status}")
    
```

```python
while True:
    student , marks = user_input()
    avg = calculate_avg(marks)
    grade = assign_grade(avg)
    status = pass_or_fail(avg)
    display_result(student, avg, grade, status)
    next = input("Do you want to add another student? yes/no")
    if next == "yes":
        continue
    else:
        break
    
```
