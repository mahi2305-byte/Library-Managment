```python
import random 

def employee_input():
    employee_name = input("Enter Employee Name")
    basic_salary = int(input("Enter baisc salary of employee"))
    year_experience = float(input("Enter years of Experience"))
    
    return employee_name , basic_salary , year_experience
    
```

```python
def calculate(basic_salary):
    hra = (20/100) * basic_salary  
    da = (15/100) * basic_salary
    e_bonus= (random.randint(5,25)/100)*basic_salary
    pf = (12/100)*basic_salary
    
    return hra , da , e_bonus , pf 
    

    
    
    
    
```

```python
def calculate_tax(gross_salary):
    if gross_salary <= 50000:
        tax = 0
    elif 50001 <= gross_salary <= 100000:
        tax = 0.1*gross_salary
    else:
        tax = 0.2 *gross_salary
        
    return tax 

    
    
```

```python
def display_salary_slip(employee_name , basic_salary , year_experience , hra , da , e_bonus , pf , gross_salary , tax , deduction , net_salary , grand_total,employee_number):
    print ("employee number" , employee_number)
    print ("Name of employee:" , employee_name)
    print( "years of experience" , year_experience)
    print(f"\n{'='*40}")
    print (f"{'SALARY SLIP' : ^40}-{employee_name}")
    print(f"{'='*40}")
    print(f"Basic salary:              ${basic_salary:,.2f}")
    print(f"HRA(20%):                  ${hra:,.2f}")
    print(f"DA(15%):                   ${da:,.2f}")
    print(f"Experience Bonus(20%):     ${e_bonus:,.2f}")
    print(f"{'-'*40}")
    print(f"Gross Salary:              ${gross_salary:,.2f}")
    print(f"Tax :                      ${e_bonus:,.2f}")
    print(f"PF(12%) :                  ${pf:,.2f}")
    print(f"{'-'*40}")
    deduction = tax + pf 
    print(f"Total Deductions:          ${deduction:,.2f}")
    print(f"{'='*40}")
    print(f"NET SALARY:                ${net_salary:,.2f}")
    print(f"{'='*40}")
    
    
 
def display_payroll_slip(total_gross , employee_number , total_deduction , grand_total):
    print(f"\n{'='*40}")
    print (f"{'PAYROLL SLIP' : ^40}")
    print(f"{'='*40}")
    
    print(f"Total Deductions:           ${total_deduction:,.2f}")
    print(f"Total Net Salary:           ${grand_total:,.2f}")
    average_net_salary = grand_total/employee_number
    print(f"Average Net Salary:           ${average_net_salary:,.2f}")
    
    print(f"{'='*40}")
    
    
    
    
    
    
    
       
    
    
    
    
    
```

```python
grand_total = 0
total_deduction = 0
total_gross = 0
employee_number = 0 


while True:
    
    name , basic_salary , year_experience = employee_input()
    hra , da , e_bonus , pf = calculate(basic_salary)
    gross_salary = basic_salary + hra + da + e_bonus 
    tax = calculate_tax(gross_salary)
    total_deduction = tax + pf 
    net_salary = gross_salary - total_deduction 
    grand_total += net_salary 
    employee_number += 1 
    
    display_salary_slip(name, basic_salary , year_experience , hra , da , e_bonus , pf , gross_salary , tax , total_deduction , net_salary , grand_total,employee_number)
    choice = input("Do you want to add another employee? yes/no")
    if choice == "yes":
        continue
    else:
        break 

display_payroll_slip(total_gross , employee_number , total_deduction , grand_total)        
```
