```python
books = [[101,"python programming",True],
              [102,"artificial intelligence",True],
              [103,"machine learning",True],
              [104,"data science",True],
              [105,"java for beginners",True],
              [106,"mathematics ",True],
              [107,"cse with python",True],
              [108,"web development",True],
              [109,"database and algorithms",True],
              [110,"python with ibm",True],
              
              ]


    
```

```python
def issue_book(book_id):
    found = False
    for book in books:
     
        if book[0] == book_id:
            found = True 
            if book[2]:
                book[2] = False 
                print ("Book issues successfully ")
                print(book)
            else :
                print("book is unavailable ")
                
    if not found :
        print("book id not found")    
    
```

```python
def return_book(book_id):
    found = False
    for book in books:
     
        if book[0] == book_id:
            found = True 
            if not book[2]:
                book[2] = True 
                print ("Book returned successfully ")
                print(book)
            else :
                print("book has not been issued ")
                
    if not found :
        print("book id not found")    
    
```

```python
def display_books():
    print("\n ID   TITLE                       STATUS")
    print("-"*50)
    for book in books:
        book_id, title , is_available = book 
        status = "Available" if is_available else "Issued"
        print(f"{book_id:<6}{title:<25}{status}")
    
```

```python
def search_book(title):
    found = False 
    for book in books:
        if title.lower() == book[1].lower():
            found = True
            status = "Available" if book[2] else "Unavailable"
            print(f"{book[0]} , Title: {book[1]} , Status : {status}")
            
    if not found :
        print ("Book is not in library")    
```

```python
def show_stats():
    total = len(books)
    available = 0
    issued = 0 
    
    for book in books:
        if book[2]:
            available += 1
        else :
            issued += 1
            

    print(f"\n{'=' * 40}")
    print(f"{'STATICS' : ^ 40}")
    print(f"{'=' * 40}")
    print(f"Total Books : {total}")
    print(f"Available : {available}")
    print(f"Issued : {issued}")
    print(f"{'=' * 40}")
    
 
    
    
```

```python
def display_menu():
    print(f"\n {'='*40}")
    print (f"{'LIBRARY MANAGEMENT SYSTEM' : ^40}")
    print(f"{'='*40}")
    print("1.Display all books")
    print("2.Issue a book")
    print("3.Return a book")
    print("4.search a book")
    print("5.Show Statistics")
    print("6.Exit")
    print(f"{'=' *40}")
```

```python
while True:
    display_menu() 
    choice = int(input("ENTER CHOICE:"))

    if choice == 1 :
        display_books()
    elif choice == 2:
        book_id = int(input("Enter Book ID to issue:"))
        issue_book(book_id)
    elif choice == 3:
        book_id = int(input("Enter Book ID to return"))
        return_book(book_id)
        
    elif choice == 4:
        title =  input("Enter Book Title to search:")
        search_book(title)
    elif choice == 5:
        show_stats()
    elif choice == 6:
        print("Exiting the system . Goodbye!")
        break 
    else:
        print("Invalid choice , enter correct choice")
        
        
        
    
```
