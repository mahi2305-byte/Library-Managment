```python
class Book:
    def __init__(self , title , author , copies):
        self.title = title 
        self.author = author 
        self.copies = copies 
        
        return title , author , copies 
    
    def borrow_book(self , copies ):
        if self.copies <= 0:
            raise ValueError("no copy available for borrowing")

        self.copies -= 1 
        print ("book issued successfully")
        print ("copies available after issuing" , copies)
        
        return self 
    
    def return_book(self , copies):
        self.copies += 1 
        print("book returned successfully")
        print ("copies available after returning" , copies )
```

```python
class Member:
    def __init__(self,name):
        self.name = name 
        self.borrowed_books = []
        
    def borrow(self , book : Book): #(why book:Book that book paramenter will have all parameter from Book class )
        if len (self.borrowed_books) >= 3:
            raise ValueError("cannot borrow more than 3 books at a time")     #difference between valueerror and print error are used to keep record of what went wrong during execution
        book.borrow_book()                                                     # raise means giving more priority of error 
        
        self.borrowed_books.append(book.title)
    
    def return_book (self , book : Book):
        if book.title not in self.borrowed_books:
            raise ValueError("this book was not borrowed by member")
        book.return_book()
        
        self.borrowed_books.remove(book.title)
```

```python
class Library:
    books = []
    members = []
    
    @classmethod
    def add_book(cls , book : Book):
        cls.books.append(book)
        
    @classmethod
    def register_member(cls, member : Member):
        cls.members.append(member)
        
    def view_books(self):
        if not self.books:
            print("books not available to display")
            
        print (f" AVAILABLE BOOKS" ^ 40)
        for book in self.books:
            print (f"Title: {book.title} ,Author: {book.author} , copies available {book.copies}")
            
        
    def generate_borrowed_books_report(self):
        print(f"Library Report" ^ 40)
        for idx , member in enumerate(Library.members):
            for book in member.borrowed_books:
                print(f"{member.name}")
                print(f"{books}")
                
            
        
        
    
    
```
